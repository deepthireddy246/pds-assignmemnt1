```python
import pandas as pd
```

```python
# importing the data from the csv file
data_frame = pd.read_csv('https://raw.githubusercontent.com/deepthireddy246/pds-assignmemnt1/main/Question2/Raw_Data/Raw_Student_Data.csv')
```

```python
data_frame
```

```python
# strip the trailing spaces
data_frame.columns = data_frame.columns.str.strip()
```

```python
#rename the file
data_frame.rename(columns = {'race/ethnicity':'race_or_ethnicity',
                     'parental level of education':'parental_level_of_education',
    'test preparation course':'test_preparation_course',
                       'math score':'math_score',
                       'reading score':'reading_score',
                        'writing score':'writing_score'}, inplace = True)
```

```python
#save the clean file to data_clean
data_frame.to_csv('https://raw.githubusercontent.com/deepthireddy246/pds-assignmemnt1/main/Question2/Clean_Data/Clean_Student_Data.csv')
```
